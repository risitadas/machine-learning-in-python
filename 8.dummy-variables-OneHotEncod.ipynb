Dummy Variables and One Hot Encoding,

1. using pandas get_dummies
2. using sklearn OneHotEncod

build a predictor function to predict price of a home,
1) with 3400 sqr ft area in west windsor
2) with 2800 sqr ft home in robbinsville

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/risitadas/machine-learning-in-python/main/townprices-csv.csv')
df
```

1. using pandas to create dummy variables

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies] , axis = 'columns')
merged
```

```python
final = merged.drop(['town'] , axis = 'columns')
final
```

Dummy Variable Trap

```python
final = final.drop(['west windsor'] , axis = 'columns')
final
```

```python
X = final.drop(['price'] , axis = 'columns')
X
```

```python
y = final.price
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
# 3400 sqr ft home in west windsor
model.predict([[3400,0,0]])
```

```python
# 2800 sqr ft home in robbinsville
model.predict([[2800,0,1]])
```

2. using sklearn OneHotEncoder

```python
# labelling the encoder to convert town names into numbers
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town' , 'area']].values
```

```python
X
```

```python
y = dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town' , OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
```

```python
X
```

```python
model.fit(X, y)
```

```python
# 3400 sqr ft home in west windsor
model.predict([[0,1,3400]])
```

```python
# 2800 sqr ft home in robbinsville
model.predict([[1,0,2800]])
```
