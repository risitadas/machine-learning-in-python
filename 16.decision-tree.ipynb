Machine Learing with Python : Decision Tree Classification

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/risitadas/machine-learning-in-python/main/salaries-csv.csv')
df
```

```python
df.head()
```

```python
inputs = df.drop('salary_more_then_100k' , axis = 'columns')
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs
```

```python
inputs.head()
```

```python
inputs_n = inputs.drop(['company','job', 'degree'], axis = 'columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```


1. is salary of Google, Computer Engineer, Bachelors degree > 100 k ?

```python
model.predict([[2,1,0]])
```

2. is salary of Google, Computer Engineer, Masters degree > 100 k ?

```python
model.predict([[2,1,1]])
```
