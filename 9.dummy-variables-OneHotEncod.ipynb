OneHotEncoding : 

required to model the price of cars with the available independent variables,

1. Price of mercedez benz that is 4 yr old with mileage 45000

2. Price of BMW X5 that is 7 yr old with mileage 86000

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/risitadas/machine-learning-in-python/main/carprices-csv.csv')
df
```

```python
df['Car Model'].value_counts()
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df, dummies], axis = 1)
merged
```

```python
final = merged.drop(['Car Model','Mercedez Benz '], axis = 'columns')
final
```

```python
X = final.drop('Sell Price' , axis = 'columns')
X
```

```python
y = final['Sell Price']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

1. Price of mercedez benz that is 4 yr old with mileage 45000


```python
model.predict([[45000,4,0,0]])
```


2. Price of BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[86000,7,0,1]])
```
