```python
# importing the essential libraries
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from word2number import w2n
```

```python
# loading the dataset
df = pd.read_csv('https://raw.githubusercontent.com/risitadas/machine-learning-in-python/main/hiring-multiple-csv.csv')
df
```

```python
# filling the values with zeroes
df.experience = df.experience.fillna("zero")
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
# calculating mean of the test score 
import math
median_test_score = math.floor(df['test_score(out of 10)'].mean())
median_test_score
```

```python
# filling the median score of test score with NaN vaues
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score)
df
```

```python
# splitting the data into X and y
# creating instance of Linear Regression
# and applying Linear Regression
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

1. predicting the salary for 2 year experience and 9 test score and 6 interview score

```python
reg.predict([[2,9,6]])
```

2. predicting the salary 12 year experience, 10 test score and 10 interview score

```python
reg.predict([[12,10,10]])
```
